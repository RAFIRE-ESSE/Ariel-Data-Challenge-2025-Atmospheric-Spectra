```python
def normalize(arr, t_min, t_max):
    shape=arr.shape
    arr=arr.reshape(-1)
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return numpy.array(norm_arr).reshape(shape)
```
